# Title: The Role of Artificial Intelligence (AI) in Efficient Healthcare Decision-Making

Timely and accurate diagnosis is foundational to good clinical practice and optimal patient outcomes. Diagnostic accuracy in healthcare refers to the extent to which tests and procedures correctly diagnose various diseases or conditions. Problems of diagnostic inaccuracy can result to missed or delayed diagnoses, leading to harm for patients.

# Statement of the Problem

Inaccurate or delayed diagnosis is currenly common in our healthcare, and represents a major public health problem likely to affect every one of us at least once in our lifetime, sometimes with devastating consequences (U.S. National Academies of Sciences, Engineering, and Medicine quality report, 2015).

# Aim and Objectives of the Knowledge Showcase Project

The aim of this project is to investigate the role of Artificial Intelligence (AI) in efficient healthcare decision-making.

The specific objectives of this study are:

1. To investigate the descriptive statistics and relationship between variables of the dataset
2. To investigate Age, Gender, and Diagnosis relationships with healthcare outcomes and patient satisfaction
3. To investigate the distribution of AI diagnosis confidence across key variables
3. To investigate the AI diagnosis confidence in relation to diagnosis and treatment duration

# Data Collection and Methodology

The data was obtained from Kaggle, and visualized using the python libraries such as Pandas, seaborn, Matplotlib.pyplot.

# Target Audience
This project will provide insights and recommendations to healthcare teams to optimize diagnosis and decision-Making in the healthcare system

# Importing the Necessary Libraries

```python
#Importing Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading / Displaying the Dataset

```python
# Loading the dataset "AI_in_HealthCare_Dataset.csv"
df = pd.read_csv('AI_in_HealthCare_Dataset.csv')

# Displaying the first five rows of the dataset
df.head()
```

# Data Exploration

```python
# checking the summary of the dataframe
df.info()
```

```python
# Checking for missing values
df.isnull().sum()
```

## Descriptive Statistics and Exploration of the Distribution of Numerical Variables and AI Diagnosis Confidence

```python
# Summary of statistics for the numerical columns
df.describe()
```

```python
# Calculate the percentage of each gender
plt.figure(figsize=(6, 6))

df['Gender'].value_counts().plot.pie(autopct='%1.1f%%', labels=['Male', 'Female'], colors=['skyblue', 'lightcoral'])
plt.title('Percentage of Male and Female')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Save picture as pdf and jpg
plt.savefig('artifact/AI_in_Efficient_Healthcare_by_Gender.pdf', format='pdf')
plt.savefig('image/AI_in_Efficient_Healthcare_by_Gender.jpg')

# plt.show displays the plot
plt.show()
```

```python
# Histogram of numerical variables
plt.figure(figsize=(12, 8))

# List of numerical variables
numerical_vars = ['Age', 'Blood_Pressure', 'Heart_Rate', 'Temperature']

for i, column in enumerate(numerical_vars, 1):
    plt.subplot(2, 3, i)
    sns.histplot(data=df, x=column, kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.savefig('artifact/AI_in_Efficient_Healthcare_Nvariable.pdf', format='pdf')
plt.savefig('image/AI_in_Efficient_Healthcare_Nvariable.jpg')
plt.show()
```

The histograms above show the treatment duration by Age, Blood_Pressure, Heart_Rate, and Temperature

# Relationship Between Age, Gender and Diagnosis with Patient Satisfaction

```python
# Bar plot of Patient Satisfaction by Age Group
plt.figure(figsize=(10, 6))

sns.barplot(data=df, x=pd.cut(df['Age'], bins=range(20, 100, 10)), y='Patient_Satisfaction')
plt.title('Patient Satisfaction by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Patient Satisfaction')
plt.xticks(rotation=45)

plt.savefig('artifact/Patient_Satisfaction_by_Age_Group.pdf', format='pdf')
plt.savefig('image/Patient_Satisfaction_by_Age_Group.jpg')
plt.show()
```

The patient satisfaction was generally high, but highest among patient of 70 to 80 years old

```python
# Box plot of Patient Satisfaction by Gender
plt.figure(figsize=(8, 6))

sns.boxplot(data=df, x='Gender', y='Patient_Satisfaction')
plt.title('Patient Satisfaction by Gender')
plt.xlabel('Gender')
plt.ylabel('Patient Satisfaction')

plt.savefig('artifact/Patient_Satisfaction_by_Gender.pdf', format='pdf')
plt.savefig('image/Patient_Satisfaction_by_Gender.jpg')
plt.show()
```

The patient satisfaction was lowest at 2/5 and highest at 4/5 for both male and female

```python
# Box plot of Patient Satisfaction by Diagnosis
plt.figure(figsize=(12, 6))

sns.boxplot(data=df, x='Diagnosis', y='Patient_Satisfaction')
plt.title('Patient Satisfaction by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Patient Satisfaction')
plt.xticks(rotation=45)

plt.savefig('artifact/Patient_Satisfaction_by_Diagnosis.pdf', format='pdf')
plt.savefig('image/Patient_Satisfaction_by_Diagnosis.jpg')
plt.show()
```

The patient satisfaction was also lowest at 2/5 and highest at 4/5 across Diagnosis

# Distribution of AI Diagnosis Confidence Across Key Variables

```python
# List of columns for subplots
columns = ['Medication', 'Insurance_Type', 'X-ray_Results', 'Surgery_Type', 'Recovery_Time', 
           'Allergies', 'Family_History', 'Patient_Satisfaction']

# Create subplots for each column
plt.figure(figsize=(11, 20))
for i, column in enumerate(columns, 1):
    plt.subplot(6, 2, i)
    sns.boxplot(data=df, x=column, y='AI_Diagnosis_Confidence', palette='Set2')
    plt.title(f'AI Diagnosis Confidence by {column}')
    plt.xlabel(column)
    plt.ylabel('AI Diagnosis Confidence')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig('artifact/Patient_Satisfaction_Across_Key_Variables.pdf', format='pdf')
plt.savefig('image/Patient_Satisfaction_Across_Key_Variables.jpg')
plt.show()
```

AI dianostic confidence in Medication, Insurance_Type, X-ray_Results, Surgery_Type, Recovery_Time, Allergies, Family_History and Patient_Satisfaction ranged from 0.77 t0 0.92 confidence levels.

# AI Diagnosis Confidence in Relation to Diagnosis and Treatment Duration

```python
# Box plot showing the distribution of AI Diagnosis Confidence for Treatment Duration in the dataset

plt.figure(figsize=(8, 6))

sns.boxplot(data=df, x='Treatment_Duration', y='AI_Diagnosis_Confidence', palette='Set2')
plt.title('AI Diagnosis Confidence by Treatment Duration')
plt.xlabel('Treatment_Duration')
plt.ylabel('AI Diagnosis Confidence')

plt.savefig('artifact/Patient_Satisfaction_by_Treatment_Duration.pdf', format='pdf')
plt.savefig('image/Patient_Satisfaction_by_Treatment_Duration.jpg')
plt.show()
```

The confidence level for treatment duration resulting from AI diagnosis ranged from 0.76 to 0.94

```python
# Box plot showing the distribution of AI Diagnosis Confidence for each Diagnosis category in the dataset
plt.figure(figsize=(8, 6))

sns.boxplot(data=df, x='Diagnosis', y='AI_Diagnosis_Confidence', palette='Set2')
plt.title('AI Diagnosis Confidence by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('AI Diagnosis Confidence')

plt.savefig('artifact/Patient_AI_Diagnosis_Confidence.pdf', format='pdf')
plt.savefig('image/Patient_AI_Diagnosis_Confidence.jpg')
plt.show()
```

AI Diagnosis Confidence revealed a higher accuracy in cancer diagnosis
